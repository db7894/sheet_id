```python
import os, sys
import numpy as np
import scipy.misc
import matplotlib.pyplot as plt
from skimage.measure import structural_similarity as ssim
import cv2
#from skimage.transform import rotate
from PIL import Image
import imagehash
#import argparse
#import shelve
import glob
```

```python
pdf = os.path.join(os.getcwd(), 'pdf')
CELLPHONE_IMG_PATH = os.path.join(pdf, 'minidataset_cell')
DB_PATH            = os.path.join(pdf, 'mini_dataset_png')
```

```python
png_paths = glob.glob(os.path.join(DB_PATH, '*.png'))
```

```python
db = {}
```

```python
testImgPath = os.path.join(CELLPHONE_IMG_PATH, 'IMG_9137.jpg')
```

```python
img = cv2.imread(testImgPath, cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap="gray")
```

```python
db = {}
```

```python
img.shape
```

```python
for i in range(len(png_paths)):
    imagePath = png_paths[i]
    image = cv2.imread(imagePath, cv2.IMREAD_GRAYSCALE)
    resized_image = cv2.resize(image, (3024,4032)) 
    
    #print(resized_image.shape)
    
    similarity = ssim(img, resized_image)
    
    filename = imagePath[imagePath.rfind("/") + 1:]
    
    db[filename] = similarity
```
