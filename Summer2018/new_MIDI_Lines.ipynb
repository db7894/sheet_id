We want all events happening at the same time. What I'll do is this:

- iterate thru MIDI messages
- every time I encounter a new time that wasn't there before, add the time to a dict / hash table
- store the message as an element with the key corresponding to its time in the dict / hash table

```python
# save the file
import pickle as pickle
def save_obj(obj, name ):
    with open(name + '.pkl', 'wb') as f:
        pickle.dump(obj, f, pickle.HIGHEST_PROTOCOL)
```

```python
def load_obj(name ):
    with open(name + '.pkl', 'rb') as f:
        return pickle.load(f)
```

```python
midi_time_dict = {}
```

```python
mid = mido.MidiFile(midi_file_path)
for msg in mid.play():
    time = msg.time
    if time not in midi_time_dict.keys():
        midi_time_dict[time] = [msg]
    else:
        print("already existing time... appending")
        midi_time_dict[time].append(msg)
```
